```r
require(cowplot)
require(data.table)
require(ggplot2)
require(ggpubr)
```

```r
empty_plot = ggplot()
```

# Figure 3

```r
plist = list(
    E=readRDS("panels_rds/fig_3e.rds"),
    F=readRDS("panels_rds/fig_3f.rds"),
    G=readRDS("panels_rds/fig_3g.rds"),
    H=readRDS("panels_rds/fig_3h.rds"),
    I=readRDS("panels_rds/fig_3i.rds"),
    J=readRDS("panels_rds/fig_3j.rds")
)
```

```r
options(repr.plot.width=20, repr.plot.height=10)
p = plot_grid(
    plot_grid(plotlist=plist[1:4], nrow=1, ncol=4, align="hv", axis="blt", labels=letters[5:8]),
    plot_grid(plotlist=plist[5:6], nrow=1, ncol=2, align="hv", axis="blt", labels=letters[9:10]),
    nrow=2
)
p
```

```r
ggsave(plot=p, file="figures/fig_3.png", width=16, height=8, bg="white")
ggsave(plot=p, file="figures/fig_3.jpg", width=16, height=8, bg="white")
```

# Figure X1

```r
plist = list(
    A=plot_grid(plotlist=lapply(1:6, function(i) { readRDS(sprintf("panels_rds/supp_fig_x1a.%d.rds", i)) }), nrow=1),
    B=plot_grid(plotlist=lapply(1:6, function(i) { readRDS(sprintf("panels_rds/supp_fig_x1b.%d.rds", i)) }), nrow=1),
    C=plot_grid(plotlist=lapply(1:6, function(i) { readRDS(sprintf("panels_rds/supp_fig_x1c.%d.rds", i)) }), nrow=1),
    D=plot_grid(plotlist=lapply(1:6, function(i) { readRDS(sprintf("panels_rds/supp_fig_x1d.%d.rds", i)) }), nrow=1)
)
```

```r
options(repr.plot.width=24, repr.plot.height=15)
p = plot_grid(plotlist=plist, nrow=4, align="hv", axis="blrt", labels="auto")
p
```

```r
ggsave(plot=p, file="figures/supp_fig_x1.png", width=24, height=15, bg="white")
ggsave(plot=p, file="figures/supp_fig_x1.jpg", width=24, height=15, bg="white")
```
